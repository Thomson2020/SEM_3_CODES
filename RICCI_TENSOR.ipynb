```python
import sympy
from einsteinpy.symbolic import RicciTensor, RicciScalar
from einsteinpy.symbolic.predefined import AntiDeSitter
from einsteinpy.symbolic.predefined import DeSitter


sympy.init_printing()


metricD = DeSitter()
metricD.tensor()
metricAD = AntiDeSitter()
metricAD.tensor()

RicD = RicciTensor.from_metric(metricD)
RicD.tensor()

R_D = RicciScalar.from_riccitensor(RicD)
R_D.simplify()
R_D.expr

Ric = RicciTensor.from_metric(metricAD)
Ric.tensor()

R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr

```

```python
import numpy as np
import matplotlib.pyplot as plt

M = 1
Lval=[20,40,60]
r = np.linspace(1, 50, 1000)

def antidesitter(M, r, L):
    f = (1 - (2 * M / r) + r**2 / L**2)
    return f

print(antidesitter(1, 10, 10))

def DeSitter(M, r, L):
    f = (1 - (2 * M / r) - r**2 / L**2)
    return f

print(DeSitter(1, 10, 10))

def schwarczhild(M, r):
    f = (1 - (2 * M / r))
    return f

print(schwarczhild(1, 10))

ads_values = antidesitter(M, r, L)
ds_values = DeSitter(M, r, L)
sch_values = schwarczhild(M, r)

plt.figure(figsize=(10, 6))
for L in Lval:
  plt.plot(r, antidesitter(M, r, L), label=f'AntiDeSitter for L={L}',linestyle='dotted')
  plt.plot(r, DeSitter(M, r, L), label=f'DeSitter for L={L}')

plt.plot(r, schwarczhild(M, r), label=f'Schwarzschild',linestyle='dashed')
plt.legend()
plt.xlabel('r')
plt.ylabel('f(r)')
plt.title('Behaviour of DeSitter, Anti-DeSitter and Schwarzschild metrics')
plt.grid()
```
